```python
import numpy as np
```

```python
# The syntax is like list with [] when entering the elements in python. 
a=np.array([5,6,9])
# When working more than one dimentional array, a parent [] need to be added. each child [] is a row. 
# in the child [] each element is separated by a comma to form columns
b=np.array([[1,2],[3,4],[5,6]])
```

```python
print(a.ndim)# dimention of array
print(a.dtype)# data type of array
print(a.size) # tottal number elements of array
print(a.shape) # shape of array
print(b.ndim)
print(b.dtype)
print(b.size)
print(b.shape) # shape of array
```

```python
c=np.array([[1,2],[3,4],[5,6]],dtype=np.float64)
print(c.ndim)
print(c.dtype)
print(c.size)
print(c.shape) 
```

```python
d=np.array([[1,2],[3,4],[5,6]],dtype=complex)
print(d)
print(d.ndim)
print(d.dtype)
print(d.size)
print(d.shape)
```

```python
# the syntax is like a tuple with () when specifying automatic structured data
e=np.zeros((3,4))
f=np.ones((3,4))
g=np.arange(1,5)
h=np.arange(1,5,1)
l=np.linspace(1,5,5)
print(e)
print(f)
print(g)
print(h)
print(l)
```

```python
print(b)
print(b.reshape(2,3))
print(b.reshape(1,6))
print(b.ravel())# make it one dimentional
```

```python
print(b.sum(axis=0))
print(b.sum(axis=1))
```

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print(a)
print(b)
```

```python
print(a+b)
print(a*b) # element-wise  multiply
print(np.multiply(a,b)) # element-wise multiply

print(a.dot(b)) # matrix multiply
```

This is slicing/stacking/indexing in numpy array: https://www.youtube.com/watch?v=_d_Ka-ks2a0&list=PLeo1K3hjS3uset9zIVzJWqplaWBiacTEU&index=3

```python
a=np.array([6,7,8])
print(a[0:2]) # slice from index 0 to 1
print(a[-1]) # the last index
```

```python
b=np.array([[6,7,8],[1,2,3],[9,3,2]])
print(b[1,2])# get elements at row 1 and column 2
print(b[0:2,2])# get elements at row 0+1 and column 2
print(b[-1]) # get the last row
print(b[-1,0:2]) # get elements at the last row and the 1st and 2nd column 
print(b[:,0:2]) # get elements of all rows and the 1st and 2nd column 
```

```python
# get each row of array
for row in b:
    print (row)
```

```python
# get each row of array, then get column elements of each row
for row in b:
    print (row)
    for cell in row:
        print(cell)
```

Iterate numpy array using nditer: https://www.youtube.com/watch?v=XawR6CjAYV4&list=PLeo1K3hjS3uset9zIVzJWqplaWBiacTEU&index=4

```python
a=np.arange(12).reshape(3,4)
print(a)
```

![alt text](image-2.png) ![alt text](image-3.png)

```python
#print each element
for x in np.nditer(a,order='C'):
    print(x)

for x in np.nditer(a,order='F'):
    print(x)
```

```python
#print each column as an array
for x in np.nditer(a,order='F',flags=['external_loop']):
    print(x)
```
